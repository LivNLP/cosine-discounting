# Load WiC dataset

```python
import pandas as pd
```

```python
data = pd.read_csv('WIC_default_V4_emb.csv')
```

```python
# shaffle all dataFrame
data = data.sample(frac=1).reset_index(drop=True)
data
```

```python
data.index
```

```python
# 10 bins split
maxy = max(data['freq'])
maxy
```

# Prepare 5 cross-validations dataset

```python
import numpy as np
```

```python
no_bin=10
move = np.log2(maxy)/no_bin
validation_index={i:[] for i in range(1,6)}
# {validation index 1 to 5 with the index of 5 validation}
total=0
for index in range(no_bin):
    left = index*move
    if index == no_bin-1:
        right = np.log2(maxy)+5
    else:
        right = (index+1)*move
    temp_index = list(data[(data['log2freq']>=left) & (data['log2freq']<right)].index)
    total+=len(temp_index)
    #print(left, right, len(temp_index))
    for validation in range(5):
        val_left=(validation)*round(len(temp_index)/5)
        val_right=(validation+1)*round(len(temp_index)/5)
        if validation!=4:
            for member in temp_index[val_left:val_right]:
                validation_index[validation+1].append(member)
        else:
            for member in temp_index[val_left:]:
                validation_index[validation+1].append(member)
        #print(validation_index)
```

```python
data.loc[validation_index[5]]
```

```python
# divide into csv 1 to 5
all_validation = [1,2,3,4,5]
for i in range(1,6): # i = 1 to 5
    test = data.loc[validation_index[i]]
    train_index=[]
    for h in all_validation:
        if h!=i:
            for member in validation_index[h]:
                train_index.append(member)
    train=data.loc[train_index]
    test.to_csv('WIC_test_val'+str(i)+'.csv')
    train.to_csv('WIC_train_val'+str(i)+'.csv')
```
